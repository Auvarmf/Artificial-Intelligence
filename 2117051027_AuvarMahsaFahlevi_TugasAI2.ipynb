## Library

```python
import pandas as pd
import numpy as np
import skfuzzy as sf
import matplotlib.pyplot as plt
```

```python
# gaji dalam usd
Sedikit = [500, 500, 1200]    # Sedikit
Sedang  = [1000, 2000, 3000]  # Sedang
Banyak  = [3000, 5000, 5000]  # Banyak
```

```python
# output range gaji dari 500 sampai 5000 dollar
usd_graph = np.arange(500, 5001) # 5001 agar output graph sampai 5000
usd_graph
```

```python
# draw member function
def drawMSGaji(usd_graph, usdset, lset):
  usd = []
  for i in usdset:
    usd.append(sf.trimf(usd_graph, i))

    fig, ax = plt.subplots()
    for member, label in zip(usd, lset):
      ax.plot(usd_graph, member, label = label)

      # output
      ax.legend()
      fig.show()
```

```python
# function set
usdset = [Sedikit, Sedang, Banyak]

# label set
lset = ["Sedikit", "Sedang", "Banyak"]
```

```python
# output grafik plot
drawMSGaji(usd_graph, usdset, lset)
```
